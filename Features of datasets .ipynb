The data set contains a total of 76 attributes out of which 14 will be used

Age( Age in years)

Sex(1=male & 0=female)

cp: chest pain --
Value 1: typical angina:substernal chest pain precipitated by physical exertion or emotional stress and relieved with rest or nitroglycerin

Value 2: atypical angina:When one experiences chest pain that doesn't meet the criteria for angina, it's known as atypical chest pain -- 

Value 3: non-anginal pain: recurrent chest pain that is indistinguishable from ischemic heart pain after a reasonable workup has excluded a cardiac cause -- 

Value 4: asymptomatic:Asymptomatic heart attacks medically known as silent myocardial infarction (SMI) annually accounts for around 45-50% of morbidities due to cardiac ailments and even premature deaths in India. The incidences among middle aged people experiencing SMI is twice likely to develop in males than females. The symptoms of SMI being very mild in comparison to an actual heart attack; it is described as a silent killer. Because sometimes there is no cardiac arrest symptoms.

Trestbps: resting blood pressure (in mm Hg on admission to the hospital)
chol: serum cholestoral in mg/dl

Fbs:fast blood sugar
level of 99 mg/dL or lower is normal, 100 to 125 mg/dL indicates you have prediabetes, and 126 mg/dL or higher indicates you have diabetes.
(1 = true; 0 = false) 

restecg: resting electrocardiographic results -- 
Value 0: normal -- 
Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) -- 
Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

thalach: maximum heart rate achieved
 
exang: exercise induced angina (1 = yes; 0 = no)

oldpeak = ST depression induced by exercise relative to rest

slope: the slope of the peak exercise ST segment -- 
Value 1: upsloping -- 
Value 2: flat -- 
Value 3: downsloping
